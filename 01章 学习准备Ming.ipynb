### <font color = red>2 简答题
#### Python 语⾔和 Julia 语⾔有什么不同？各有什么特点？


- julia是最初为科学计算而设计的动态编程语言，不需要解释器，运行速度快，允许并发、并行和分布式计算，并直接调用C和Fortran库，不适用粘合代码，Julia的第三方库都是用Julia语言编写；
- 相反python作为解释型语言，其强调代码可读性和语法的简洁，被普遍用作链接其他语言所编写的软件组件的“胶水语言”。

有人认为Python是一种折中方案，对数据分析工作的某些部分来说是好的，但对其他部分来说效率有限，因而julia旨在解决Python和其他语言以及科学计算和数据处理所使用的应用程序的缺点，是一种试图取其精华弃其糟粕的语言。

#### 官方↓

就差异而言,

- Python 是一种通用语言，具有科学领域的庞大用户社区和杰出的科学生态系统
- Python 的优点是得到大量科学库集合的支持(并且是一项非常有市场价值的技能and being a highly marketable skill)
- Julia是一种较新的语言，具有许多令人兴奋的特点
- 朱莉娅的优势在于第三方库通常完全用朱莉娅本身来编写。


相同点：
- 它们都是现代的、开源的、高生产率的语言，具有高性能计算所需的所有关键特性。

### <font color = red>3 简答题
使⽤开源软件 (open source software) 有哪些好处？

开源软件其源代码可以任意获取，并在软件协议规定之下允许其他使用者学习及修改。一般情况下使用者对软件运行具有控制权，因此降低了由于病毒或木马等来历不明软件的运行损害电脑系统的风险。

#### 官方↓

1. 开源意味着这意味着我们可以在我们的所有计算机上安装它们（指python等开源语言）和它们的库，而无需支付费用或考虑许可证问题。

2. 开源库的另一个优势是你可以阅读它们并了解它们是如何工作的。

3. 另外，开源软件的一个哲学优势是它符合可再生性的科学理想（scientific ideal of reproducibility.）

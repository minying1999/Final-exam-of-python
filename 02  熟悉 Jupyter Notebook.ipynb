#### <font color = red>5 简答题
“Jupyter Notebook 采⽤模态编辑 (modal editing) ⽅式” 该怎样理解？有何好处？

理解：不同的模式键盘输入实现的效果不同

1. 编辑模式
    - 单元格边框为绿色，单元格内有光标闪烁
    - 键入的任何内容都将显示在该单元格cell中（通常是代码和备注），按下运行键可以运行输入的代码
    
2. 命令模式
    - 单元格边框为灰色(或灰色和蓝色)，无光标闪烁
    - 按键被解释为命令：例如，键入 b 会在当前单元格的下面添加一个新单元格
    
很大程度上可以理解为编辑模式是对单元格内部进行操作，而命令模式是对整个jupyter notebook进行操作

### 10 上机题：Exercise 1
完成教材 “Setting up Your Python Environment” 节末尾的 Exercise 1，从⽽了解终端、Jupyter
Notebook 和 浏览器之间的关系。

- Jupyter notebook 是一个从（电脑）终端启动的网页服务进程，可以把他理解为一个运行在电脑上的网页/网站服务器（localhost），我们浏览器内看到的网页一般的是浏览器解析内核数据之后呈现给我们的。
- 浏览器关闭丝毫不妨碍 notebook 服务的运行，就像即使我们不打开某个网站，那个网站的服务器也不会关闭，而是静静地启动着、等待着我们的访问。重新打开浏览器重新访问同样的地址，运行状态都在那里，只要不关闭终端，我们可以随时通过网址访问localhost。
- --no-browser 选项是只启动服务进程，不启动浏览器，和正常的网页服务器一样，配置好服务器的网络访问权限，将可以从其他计算机 (甚至没有安装 Python 的计算机) 访问服务器上的 Jupyter notebook 服务。

简而言之，Jupyter是本地终端的一个服务进程，浏览器是我们访问终端和进程的窗口，就类似我们访问普通网页的服务器一般，由浏览器将服务器（终端）传来的进程（jupyter）数据解析给我们看

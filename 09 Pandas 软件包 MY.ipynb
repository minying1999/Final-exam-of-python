### 12 简答题
给定⼀个 DataFrame 类型的变量 df ，代表着⼀个类似于 Excel 的数据表格，其两个数据列的名称分
别是 population 和 total GDP 。population 列的数据是各个国家的⼈⼝数，单位为
“⼈”； total GDP 列的数据是各个国家的 GDP 总量，单位为 “百万美元”。如果要计算各个国家的⼈均
GDP，单位为 “美元/⼈”，将计算出的数据保存为 df 中的 GDP percap 列，应该键⼊什么代码？

```python
df['GDP percap'] = df['total GDP'] * 1e6 / df['population']
df
```

### 13 简答题
给定⼀个 DataFrame 类型的变量 df ，代表着⼀个类似于 Excel 的数据表格，其 GDP percap 列的数
据代表着各个国家的⼈均 GDP。如果希望根据这⼀列数据绘制柱状图 (bar chart)，应该键⼊什么代
码？

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
ax = df['GDP percap'].plot(kind='bar')
ax.set_xlabel('country', fontsize=12)
ax.set_ylabel('GDP per capita', fontsize=12)
plt.show()
```

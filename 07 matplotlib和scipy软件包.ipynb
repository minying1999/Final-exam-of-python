###  <font color = red>10 简答题<font size = 0.001>Ming

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.stats import norm
from random import uniform
fig, ax = plt.subplots()
x = np.linspace(-4, 4, 150)
#type(x)
for i in range(4):
    m, s = uniform(-1, 1), uniform(1, 2)
    y = norm.pdf(x, loc=m, scale=s)
    current_label = f'$\mu = {m:.2}$'
    ax.plot(x, y, linewidth=2, alpha=0.6, label=current_label)
ax.legend()
plt.show()
```

请简要介绍以上代码运⾏后将发⽣什么效果

#### 答1

先输入库
然后画了个图，x轴从-4到4，有150个样本
在for循环里生成三组数据，m服从从-1到1的均匀分布，s服从从1到2的均匀分布，y是正态分布均值是m标准差是s，图例是μ=m
画线，线宽2，透明度60%

#### 答2
这段代码运行后会按顺序执行：

先导入所需要的库，然后产生一个从 -4 到 4  的均匀等差数组x

然后开启循环，分别模拟三个正态分布函数，并绘制图例，设置细节。循环结束后用plt.legend()将三条正态分布函数的曲线绘制进同一个坐标轴，最后展示图片

运行结果是在同一个坐标轴里面画出三个图例，这三个图例都是
服从~N(m,s)分布的正态分布（概率密度）函数曲线，其中m是从`(-1.1）`之间产生的的服从均匀分布的随机数，s是`(1，2）`区间内产生的服从均匀分布的随机数。

且这三条曲线分别用它们的均值产生一个标签，并且自动为它们作区分（自动填曲线颜色和标签颜色）



###  <font color = red>11 简答题

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.stats import norm
from random import uniform
num_rows, num_cols = 3, 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 12))
for i in range(num_rows):
    for j in range(num_cols):
        m, s = uniform(-1, 1), uniform(1, 2)
        x = norm.rvs(loc=m, scale=s, size=100)
        axes[i, j].hist(x, alpha=0.6, bins=20)
        t = f'$\mu = {m:.2}, \quad \sigma = {s:.2}$'
        axes[i, j].set(title=t, xticks=[-4, 0,4], yticks=[])
plt.show()
```

答：
- 第i行第j列的图用直方图的形式画，以生成的100个服从正态分布（~N(loc,scale)）的随机数
按照频率划分成20组，展示出的图例透明度为60%

###  <font color = red>15 简答题

```python
def bisect(f, a, b, tol=10e-5):
    """
    Implements the bisection root finding algorithm, assuming that f is a
    real-valued function on [a, b] satisfying f(a) < 0 < f(b).
    """
    lower, upper = a, b
    while upper - lower > tol:
        middle = 0.5 * (upper + lower)
        if f(middle) > 0: # root is between lower and middle
            lower, upper = lower, middle
        else: # root is between middle and upper
            lower, upper = middle, upper
    return 0.5 * (upper + lower)
```

请根据以上代码简要介绍⽤⼆分法 (bisection) 寻找单调递增函数$ f$在 $(a,b)$区间的根 (root)$x$ 
满⾜$f(x) = 0 , x \in (a,b)$的算法思想是什么。

#### 答

在区间（a,b）之间寻找单挑递增函数的根时（假设a为区间左端点），如果a,b之间的距离大于所给的精确度（位置参数默认为10e-5），则计算a，b的中点middle的值，并计算f（middle）的值，由于根据条件已知跟的两侧取得点的函数值必然一正一负，所以当中点值大于零（小于零）的时候，跟必然在左端点（右端点）和中点之间，所以将中点定为新区间的右端点（左端点），左端点（右端点）保持不变，对新区间同样取新的中点，计算新的中点的值....如此反复，直到区间端点的距离小于所给的精确度（本例默认为10e-5），结束循环得出此时区间中点对应的值。这个值就是我们的对根的估计值



思想就是通过循环迭代缩小根的存在范围，当范围小于所给精确度时，此时范围区间中点的值可以看做是对根的逼近

###  <font color = red>17 简答题
简要说明⽜顿法的算法思想原理，并推导写出最关键的寻根迭代公式。

[牛顿法、拟牛顿法原理](https://blog.csdn.net/zgcr654321/article/details/89674713)

[牛顿迭代法](https://baike.baidu.com/item/%E7%89%9B%E9%A1%BF%E8%BF%AD%E4%BB%A3%E6%B3%95/10887580?fromtitle=%E7%89%9B%E9%A1%BF%E6%B3%95&fromid=1384129&fr=aladdin)

#### 答：

1. Newton-Raphson（牛顿-拉夫逊（拉弗森））方法利用切线是曲线的线性逼近的事实和单根附近具有平方收敛的性质，使用局部斜率信息来尝试提高收敛速度
    - 先假设x0为根x`*`的初始近似值，计算f（x）在x0处的线性近似函数（切线，因为有斜率有函数值，还有x0相当于可以定下来切线），然后找到切线和X轴的截距x1,此时x1已经比x0贴近x`*`了；如此反复迭代下去，计算到最后的xn已经无限接近于x`*`了，而且收敛到x`*`的效率比较高

- 或者说在当前位置x0对f（x）做二阶泰勒展开，进而找到下一个位置x的估计值。

#### 推导写出最关键的寻根迭代公式。

$$ y = f^{'}(x_0)(x-x_0)+f(x)$$

$$ 0 = f^{'}(x_0)(x-x_0)+f(x_0)$$

$$ x_1 = x_0 - \frac{f(x_0)}{f^{'}(x_0)}$$

$$x_{n+1} = x_n - \frac{f(x_n)}{f^{'}(x_n)}$$

直到$f(x_{n+1})$近似为零,迭代结束，根为$x_{n+1}$

###  <font color = red>18 简答题
请简要评价⽜顿法 (Newton's method) 寻根算法 (root-finding algorithm) 相⽐于⼆分法 (bisection) 的
优缺点。

### 答：

1. 牛顿法的主要优点是收敛速度很快，但它并不能保证收敛到根，而二等分法和割线方法都保证可以收敛到一个解（假设所涉的区间确实有根）。

2. 牛顿法还需要计算所讨论函数的导数的值，而求导的效率和函数有关，可能效率会极低

3. 牛顿法的收敛标准不同于二等分法和割线法。在这些方法中，我们知道我们与解决方案有多近，因为我们正在计算包含解决方案的间隔。用牛顿的方法，我们不知道我们离解决方案有多近，因为我们只能计算得到$f(x)$
4. 不能保证方法会收敛到解决方案，因而应该设置最大迭代次数，以便在找不到解决方案的情况下得到一个可以接受的结果。
